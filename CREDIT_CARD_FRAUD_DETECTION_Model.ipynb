### Data Pre-processing

Splitting the dataset into training and test data sets.

```python
from sklearn.model_selection import train_test_split
X = df.drop(['Class'], axis=1)
y = df['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=100)
```

# MODEL BUILDING

### LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
folds = KFold(n_splits=5, shuffle=True, random_state=4)

# params
params = {"C": [0.01, 0.1, 1, 10, 100, 1000]}

model_cv = GridSearchCV(estimator = LogisticRegression(),
                        param_grid = params,
                        scoring= 'roc_auc',
                        cv = folds,
                        verbose = 1,
                        return_train_score=True)

# Fit the model
model_cv.fit(X_train, y_train)
```

```python
cv_results = pd.DataFrame(model_cv.cv_results_)
cv_results
```

```python
cv_results.columns
```

```python
cv_results.dropna(axis=1)
```

```python
# plot of C versus train and test scores

plt.figure(figsize=(8, 6))
plt.plot(cv_results['param_C'], cv_results['mean_test_score'])
plt.plot(cv_results['param_C'], cv_results['mean_train_score'])
plt.xlabel('C')
plt.ylabel('roc_auc')
plt.legend(['test result', 'train result'], loc='upper left')
plt.xscale('log')
```

```python
best_score = model_cv.best_score_
best_C = model_cv.best_params_['C']

print(" The highest test roc_auc {0} is at C = {1}".format(best_score, best_C))
```

```python
logistic_imb = LogisticRegression(C=0.01)
logistic_imb_model = logistic_imb.fit(X_train, y_train)
y_train_pred = logistic_imb_model.predict(X_train)
confusion = metrics.confusion_matrix(y_train, y_train_pred)
print(confusion)
```

```python
TP = confusion[1,1] # true positive
TN = confusion[0,0] # true negatives
FP = confusion[0,1] # false positives
FN = confusion[1,0] # false negatives
```

```python
print("Accuracy:-",metrics.accuracy_score(y_train, y_train_pred))
print("Sensitivity:-",TP / float(TP+FN))
print("Specificity:-", TN / float(TN+FP))
print("F1-Score:-", f1_score(y_train, y_train_pred))

#Classified report
print(classification_report(y_train, y_train_pred))
```

```python
y_test_pred = logistic_imb_model.predict(X_test)
confusion = metrics.confusion_matrix(y_test, y_test_pred)
print(confusion)
TP = confusion[1,1] # true positive
TN = confusion[0,0] # true negatives
FP = confusion[0,1] # false positives
FN = confusion[1,0] # false negatives
```

```python
#function for ROC curve draw method
def draw_roc( actual, probs ):
    fpr, tpr, thresholds = metrics.roc_curve( actual, probs,
                                              drop_intermediate = False )
    auc_score = metrics.roc_auc_score( actual, probs )
    plt.figure(figsize=(5, 5))
    plt.plot( fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score )
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()

    return None
```

```python
y_train_pred_proba = logistic_imb_model.predict_proba(X_train)[:,1]
draw_roc(y_train, y_train_pred_proba)
```

```python
y_test_pred_proba = logistic_imb_model.predict_proba(X_test)[:,1]
draw_roc(y_test, y_test_pred_proba)
```

We can see that we have very good ROC on the test set 0.97, which is almost close to 1.

**Handling data imbalance**

We will test a number of methods for controlling data imbalance as we can see that it is severely out of balance.


1) Undersampling: In this case, the number of non-fraudulent transactions will be lowered to 396 (corresponding to the number of fraudulent transactions) in order to balance the class distribution.

2) Oversampling: In this case, we will count both fraudulent and non-fraudulent transactions equally.

3) SMOTE stands for synthetic minority oversampling. It is an additional oversampling method that generates fake data using the closest neighbour algorithm.

#### CONCLUSION

High precision is critical for banks with lower average transaction values since we want to correctly identify relevant fraudulent activities. This can be accomplished by incorporating a human verification step for each flagged transaction. When precision is low, however, this activity becomes demanding since more human resources are required.

Low recall, or the inability to detect fraudulent transactions, can result in considerable losses for banks with higher transaction values. As a result, in order to effectively detect actual fraudulent transactions and thereby prevent large-value fraudulent transactions, a strong recall is critical.

After assessing numerous models, it was discovered that the simplest Logistic Regression model, trained on a balanced dataset using the SMOTE technique, had a high recall and a respectable ROC score. Furthermore, it is simpler to interpret and explain to the business. As a result, this model was chosen as the best option.
